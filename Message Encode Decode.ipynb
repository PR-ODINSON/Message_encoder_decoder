```python
import tkinter as tk
from tkinter import messagebox
import base64

# Functions to encode and decode
def encode_message():
    try:
        key = key_entry.get()
        message = message_entry.get("1.0", tk.END).strip()
        
        if not key or not message:
            raise ValueError("Key and message cannot be empty.")
        
        # Encoding the message
        encoded_bytes = base64.b64encode("".join([chr(ord(char) + len(key)) for char in message]).encode())
        encoded_message = encoded_bytes.decode()
        
        result_entry.delete("1.0", tk.END)
        result_entry.insert("1.0", encoded_message)
    except Exception as e:
        messagebox.showerror("Error", str(e))

def decode_message():
    try:
        key = key_entry.get()
        encoded_message = message_entry.get("1.0", tk.END).strip()
        
        if not key or not encoded_message:
            raise ValueError("Key and encoded message cannot be empty.")
        
        # Decoding the message
        decoded_bytes = base64.b64decode(encoded_message.encode())
        decoded_message = "".join([chr(ord(char) - len(key)) for char in decoded_bytes.decode()])
        
        result_entry.delete("1.0", tk.END)
        result_entry.insert("1.0", decoded_message)
    except Exception as e:
        messagebox.showerror("Error", str(e))

# GUI setup
app = tk.Tk()
app.title("Message Encode-Decode")
app.geometry("500x400")
app.resizable(False, False)

# Labels
tk.Label(app, text="Message Encode-Decode", font=("Arial", 16, "bold")).pack(pady=10)
tk.Label(app, text="Enter your key:").pack(anchor="w", padx=20, pady=(10, 0))
key_entry = tk.Entry(app, width=50, show="*")
key_entry.pack(padx=20)

tk.Label(app, text="Enter your message:").pack(anchor="w", padx=20, pady=(10, 0))
message_entry = tk.Text(app, height=5, width=50)
message_entry.pack(padx=20)

# Buttons
button_frame = tk.Frame(app)
button_frame.pack(pady=10)

encode_button = tk.Button(button_frame, text="Encode", command=encode_message, bg="lightblue")
encode_button.pack(side="left", padx=10)

decode_button = tk.Button(button_frame, text="Decode", command=decode_message, bg="lightgreen")
decode_button.pack(side="left", padx=10)

# Result section
tk.Label(app, text="Result:").pack(anchor="w", padx=20, pady=(10, 0))
result_entry = tk.Text(app, height=5, width=50, state="normal")
result_entry.pack(padx=20)

# Run the application
app.mainloop()
```
